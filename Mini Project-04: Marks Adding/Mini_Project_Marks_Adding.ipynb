```python
import pandas as pd
```

```python
df = pd.read_csv('Mini Project - Marks Adding.csv')
df
```

```python
d =  {'Roll Num':[],"Marks":[]}
for row in df.index:
  # for each row
  roll = df['Roll Num'][row]
  marks = df['Marks'][row]
  if roll in d['Roll Num']:
    index = d['Roll Num'].index(roll)
    d['Marks'][index] += marks
  else:
    d['Roll Num'].append(roll)
    d['Marks'].append(marks)
```

```python
print(d)
```

```python
newDf = pd.DataFrame(d)
newDf
#
```

```python
newDf.to_csv("output.csv")
```
